```python
import numpy as np
import pandas as pd
import tensorflow as tf
import autokeras as ak

RANDOM_SEED = 44

df = pd.read_parquet("") # input data here
df = df.drop(['pupil_timestamp', 'start_timestamp', 'end_timestamp', 'trigger'], axis=1)
df.head()
```

```python
df.head()
```

```python
# pick Valence and Arousal as target variables
y = df[['arousal_rating']]
# y = df[['valence_rating']]
X = df.drop(['valence_rating', 'arousal_rating'], axis=1)

# print shape of X and y
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

```python
from sklearn.model_selection import train_test_split

# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.01, random_state=RANDOM_SEED)

# print shapes of train and test sets
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
automl = ak.StructuredDataRegressor(overwrite=True, max_trials=3)
```

```python
automl.fit(X_train, y_train, epochs=30)
```

```python
# Just to get smaller test size

# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_test, y_test, train_size=0.9, random_state=RANDOM_SEED)

# print shapes of train and test sets
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
import sklearn.metrics
# print statistics about the model
y_pred = automl.predict(X_test)
print("MSE score: ", sklearn.metrics.mean_squared_error(y_test, y_pred))
print("MAE score: ", sklearn.metrics.mean_absolute_error(y_test, y_pred))
print("Mean absolute percentage error score: ", sklearn.metrics.mean_absolute_percentage_error(y_test, y_pred))
```
